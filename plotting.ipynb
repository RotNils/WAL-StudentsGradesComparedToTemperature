```python
# Violin Plot

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
df = pd.read_csv("european_semester_avg.csv")

# Set style
sns.set_theme(style="whitegrid")

# Create violin plot
plt.figure(figsize=(8, 5))
sns.violinplot(x="European Semester", y="Average Numeric Grade", data=df, palette={"SoSe": "orange", "WiSe": "blue"})

# Labels and title
plt.title("Distribution of Average Grades by Semester (SoSe vs. WiSe)")
plt.ylabel("Average Numeric Grade")
plt.xlabel("Semester")

# Show plot
plt.show()
```

```python
# Box Plot

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
df = pd.read_csv("european_semester_avg.csv")

# Set style
sns.set_theme(style="whitegrid")

# Create boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x="European Semester", y="Average Numeric Grade", data=df, palette={"SoSe": "orange", "WiSe": "blue"})

# Labels and title
plt.title("Distribution of Average Grades by Semester (SoSe vs. WiSe)")
plt.ylabel("Average Numeric Grade")
plt.xlabel("Semester")

# Show plot
plt.show()
```

```python
# Histogram

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load data
df = pd.read_csv("european_semester_avg.csv")

# Set style
sns.set_theme(style="whitegrid")

# Create histogram with KDE (Kernel Density Estimate)
plt.figure(figsize=(8, 5))
sns.histplot(df, x="Average Numeric Grade", hue="European Semester", bins=20, kde=True, palette={"SoSe": "orange", "WiSe": "blue"}, alpha=0.6)

# Labels and title
plt.title("Histogram of Average Grades (SoSe vs. WiSe)")
plt.xlabel("Average Numeric Grade")
plt.ylabel("Count")
plt.legend(title="Semester")

# Show plot
plt.show()
```
